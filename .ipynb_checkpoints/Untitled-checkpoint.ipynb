```python
import requests
import re
page = requests.get("https://weather.com/weather/tenday/l/USCA0550:1:US")

from bs4 import BeautifulSoup
soup=BeautifulSoup(page.content,"html.parser")
```

```python
all=soup.find("div",{"class":"locations-title ten-day-page-title"}).find("h1").text
```

```python
print(all)
```

```python
table=soup.find_all("table",{"class":"twc-table"})
d = {}  

for items in table:
 for i in range(len(items.find_all("tr"))-1):

  #x =items.find_all("span",{"class":"day-detail"})[i].text
  d[i] = {}  
    
  var = items.find_all("span",{"class":"day-detail clearfix"})[i]
  d[i]["date"]= items.find_all("span",{"class":"day-detail clearfix"})[i].text
  d[i]["description"]=items.find_all("td",{"class":"description"})[i].text 

  var = items.find_all("td",{"class":"temp"})[i]
  d[i]["temp_high"]=re.sub('[°]', '', var.find_all("span")[0].text) 

  d[i]["temp_low"]=re.sub('[°]', '', var.find_all("span")[2].text)
  d[i]["precip"]=items.find_all("td",{"class":"precip"})[i].text
  d[i]["wind"]=items.find_all("td",{"class":"wind"})[i].text  
  d[i]["humidity"]=items.find_all("td",{"class":"humidity"})[i].text 
```

```python
del d[0]
for i in d:
    print(d[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import re

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
  
d.values()    
temp_low =  [ float((x['temp_low'])) for x in d.values() ]
temp_high = [ float((x['temp_high'])) for x in d.values() ]
for i in range(2,15):
    ax1.plot(temp_low,label=i)
    ax1.plot(temp_high,label=i)
plt.plot([0, 13], [36, 36], 'k-', color="blue")

ax1.set_xlabel('day')
ax1.set_ylabel('temp')
```
